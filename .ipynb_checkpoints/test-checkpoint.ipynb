```python
"""
# TEST PCA

 Toto je test metody PCA (principal component analysis). Zmyslom skriptu je osahat si metodu a na 
 jednoduchom priklade ju pokusit sa pochopit.
 
 Zdroje:
 
 [1] https://www.datacamp.com/tutorial/principal-component-analysis-in-python
"""



```

```python
# 1. Nacitanie vzorovych dat

from sklearn.datasets import load_breast_cancer
breast = load_breast_cancer()

breast_data = breast.data
breast_labels = breast.target
```

```python
# 2. Uprava dat podla vzoru z [1]

import numpy as np

labels = np.reshape(breast_labels,(569,1))
final_breast_data = np.concatenate([breast_data,labels],axis=1)

```

```python
# 3. Vyroba dataframe kontajnera

import pandas as pd
breast_dataset = pd.DataFrame(final_breast_data)

features = breast.feature_names

print(features)

features_labels = np.append(features,'label')

breast_dataset.columns = features_labels
breast_dataset.head()

breast_dataset['label'].replace(0, 'Benign',inplace=True)
breast_dataset['label'].replace(1, 'Malignant',inplace=True)

breast_dataset.tail()
```

```python
# 4. Standardizacia dat, t.j. z-transformace
from sklearn.preprocessing import StandardScaler

x = breast_dataset.loc[:, features].values
x = StandardScaler().fit_transform(x) # normalizing the features

print(np.mean(x),np.std(x))
```

```python
# 5. Normalizacia celeho dataframe

feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalised_breast = pd.DataFrame(x,columns=feat_cols)
normalised_breast.tail()
```

```python
# 6. PCA

from sklearn.decomposition import PCA

pca_breast = PCA(n_components=4)

principalComponents_breast = pca_breast.fit_transform(x)

principal_breast_Df = pd.DataFrame(data = principalComponents_breast
             , columns = ['pc1', 'pc2', 'pc3', 'pc4'])

principal_breast_Df.tail()

```

```python
print('Explained variation per principal component: {}'.format(pca_breast.explained_variance_ratio_))
```

```python
# PLOT
import matplotlib.pyplot as plt
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('pc3',fontsize=20)
plt.ylabel('pc4',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
targets = ['Benign', 'Malignant']
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = breast_dataset['label'] == target
    plt.scatter(principal_breast_Df.loc[indicesToKeep, 'pc3']
               , principal_breast_Df.loc[indicesToKeep, 'pc4'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```
